```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('data.csv', na_values=[ ".."], index_col=['Country Name', 'Series Name'])
data
```

```python
pt_data = data.pivot_table(data,  index=['Country Name', 'Series Name'])
```

```python
#data.dropna(axis='rows', thresh=10)
```

```python
df1 = data['2022']
#restting the index
df = df1.reset_index('Series Name')

# Filter data for the year 2022 and the specified series name
filtered_data = df[(df['Series Name'] == 'Government expenditure on education, total (% of government expenditure)')]

# Sort data by the '2022' column in descending order and select the top 10 rows
top_10_countries = filtered_data.sort_values(by='2022', ascending=False).head(10)
```

```python
# Creating a subplot with a single plot (ax[0])
fig, ax = plt.subplots()

# Plotting the data for the top 10 countries
ax.bar(top_10_countries.index, top_10_countries['2022'])
# Setting labels and title
ax.set_xlabel('Country')
ax.set_title('Top 10 Countries with Maximum Education Expenditure in 2022')

# Rotating x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

# Show the plot
plt.show()
```
