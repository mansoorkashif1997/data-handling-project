```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('data.csv', na_values=[ ".."], index_col=['Country Name', 'Series Name'])
data
```

```python
pt_data = data.pivot_table(data,  index=['Country Name', 'Series Name'])
```

```python
#data.dropna(axis='rows', thresh=10)
```

```python
df = data['2022']
#restting the index
df1 = df.reset_index('Series Name')
# Filter data for the year 2022 and the specified series name
plot1 = df1[(df1['Series Name'] == 'Government expenditure on education, total (% of government expenditure)')]

# Sort data by the '2022' column in descending order and select the top 10 rows
plot1 = plot1.sort_values(by='2022', ascending=False).head(10)
plot1 = plot1.sort_values(by='2022')
print(plot1)
```

```python
# Assuming 'Country' is a column in your DataFrame
selected_countries = ['Europe', 'South America', 'North America','Asia','Africa','Australia']

df2= data.fillna(method='ffill', axis=1)

# Filter data for the year 2022 and the specified series name
df2 = df2['2022']
# Reset the index
df2 = df2.reset_index()
# Use boolean indexing to select specific countries
df2 = df2[df2['Country Name'].isin(selected_countries)]
plot2 = df2.set_index('Country Name')
plot2 = plot2[(plot2['Series Name'] == 'Pupil-teacher ratio, primary')]
# Display the resulting DataFrame
print(plot2)
```

```python
df3= data.reset_index('Series Name')
# Filter data for the year 2022 and the specified series name
plot3 = df3[(df3['Series Name'] == 'Primary education, pupils')]
plot3 = plot3.sort_values(by='2022', ascending=False).head(3)
plot3 = plot3.dropna(axis='columns', how='all')
plot3 = plot3.fillna(method='bfill', axis=1)
print(plot3)



```

```python
df4= data.fillna(method='ffill', axis=1)

# Filter data for the year 2022 and the specified series name
df4 = df4['2022']
df4 = df4.reset_index('Series Name')

plot4 = df4[(df4['Series Name'] == 'Literacy rate, adult total (% of people ages 15 and above)')]

# Sort data by the '2022' column in descending order and select the top 10 rows
plot4 = plot4.sort_values(by='2022', ascending=False).head(7)
plot4 = plot4.sort_values(by='2022')

print(plot4)
```

```python
#Create a figure and subplots
fig, ax = plt.subplots(3, 2, figsize=(24, 18), gridspec_kw={'width_ratios': [1, 1],'height_ratios': [1, 1,1]})
fig.suptitle('''World Education Report 2022
      Name: Mansoor Kashif
        ID: 22085607''',
             fontsize=22,
             fontweight='bold')

# Plotting the data for the top 10 countries
ax[0,0].barh(plot1.index, plot1['2022'], color=['blue','yellow'])
# Setting and title
ax[0,0].set_title('Government expenditure on education (% of government expenditure ): 2022',fontsize=19,fontweight='bold' ,color='red')
ax[0,0].spines[['top', 'right', 'bottom','left']].set_visible(False)
ax[0,0].set_xlim(left=18, right=25)
ax[0,0].tick_params(labelsize=18)
ax[0,0].set_xticks([])
for i, v in enumerate(plot1.values[:, 1]):
    ax[0,0].text(v, i, '{:.2f}'.format(v), ha='left', va='center',
             fontsize=16)


ax[0,1].pie(plot2['2022'], labels=plot2.index, autopct='%1.1f%%', startangle=140)
ax[0,1].set_title('Pupil-teacher ratio', fontsize=19,fontweight='bold' ,color='red')



# Plotting the data for the top 10 countries
ax[1,0].plot(plot3.iloc[:, 1:].T)
ax[1,0].set_title('Primary education, pupils', fontsize=19,fontweight='bold' ,color='red')
ax[1,0].spines[['top','right']].set_visible(False)
ax[1,0].tick_params(labelsize=18)
ax[1,0].legend(labels=plot3.index,
           loc='center right',
           frameon=False,
           ncols=2, fontsize=18)  

# Plotting the data for the top 10 countries
ax[1,1].barh(plot4.index, plot4['2022'], color=['green','orange'])
# Setting and title
ax[1,1].set_title('Literacy rate (% of people ages 15 and above)',fontsize=19,fontweight='bold' ,color='red')
ax[1,1].spines[['top', 'right', 'bottom','left']].set_visible(False)
ax[1,1].set_xlim(left=99.5, right=100.1)
ax[1,1].tick_params(labelsize=18)
ax[1,1].set_xticks([])
for i, v in enumerate(plot4.values[:, 1]):
    ax[1,1].text(v, i, '{:.2f}'.format(v), ha='left', va='center',
             fontsize=16)


plt.show()
```
